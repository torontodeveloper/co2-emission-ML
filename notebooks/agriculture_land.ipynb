```python
import pandas as pd

# Replace 'file.csv' with your CSV file's path
file_path = "/Users/akshaysapra/co2-emission-ML/src/data/agricultural-land.csv"

# Read the CSV file into a Pandas DataFrame
data = pd.read_csv(file_path)

refined_data = data [data ['Year']>=2000]

print(refined_data.head(20))
refined_data.shape
refined_data.to_csv('agricultural-land_2000to2020.csv',index=False)
```

```python
# Load the CSV data into a DataFrame
csv_file = 'country_CO2_and_energy_2000to2018.csv'
csv_data = pd.read_csv(csv_file)

# Assuming 'refined_data' is your DataFrame
# Perform the merge based on 'Entity' and 'Year' with 'country' and 'year'
merged_data = pd.merge(refined_data, csv_data, left_on=['Entity', 'Year'], right_on=['country', 'year'], how='inner')

# Now 'merged_data' contains the merged information
# You can perform further operations or analysis on this merged DataFrame
merged_data
```

```python
merged_data.to_csv('merged_data.csv',index=False)
```
